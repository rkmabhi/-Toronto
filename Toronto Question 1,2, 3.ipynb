Creating a beautiful soup object

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

Creating an empty list to store the values. 
Finding the table from the soup object.
Creating an empty dictionary called 'cell' to store the keys and the corresponding values extracted from the table and then appending
it to the empty list that we have created.

```python
table_contents=[]
table=soup.find('table')
cell = {}
```

```python
for rows in table.find_all('td'):
    if rows.span.text =='Not assigned':
        pass
    else:
        cell['PostalCode'] = rows.p.text[:3]
        cell['Borough'] = (rows.span.text).split('(')[0]
        cell['Neighborhood'] = (((((rows.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

```python
print(table_contents)
df = pd.DataFrame(table_contents)
```
